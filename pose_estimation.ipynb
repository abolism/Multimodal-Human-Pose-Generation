```python
# Import MediaPipe and OpenCV
import mediapipe as mp
import cv2

```

```python
# Create a Pose object
mp_pose = mp.solutions.pose
pose_detector = mp_pose.Pose(
    model_complexity=2,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5)

```

```python
import matplotlib.pyplot as plt
# Create a DrawingSpec object
mp_drawing = mp.solutions.drawing_utils
drawing_spec = mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=2)

# Load an image
image = cv2.imread('tony2.jpg')
# Convert it to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# img = mpimg.imread('test.jpg')
imgplot = plt.imshow(image)
plt.show()
```

```python
# Process the image and get the pose landmarks
results = pose_detector.process(image)
pose_landmarks = results.pose_landmarks

```

```python
# Plot the pose on the image
if pose_landmarks:
  mp_drawing.plot_landmarks(
      landmark_list=pose_landmarks,
      connections=mp_pose.POSE_CONNECTIONS,
      landmark_drawing_spec=drawing_spec,
      connection_drawing_spec=drawing_spec)

# Show the image
cv2.imshow('Pose', image)
cv2.waitKey(0)
```

```python
# Create a holistic object
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils

holistic = mp_holistic.Holistic(enable_segmentation=True, min_detection_confidence=0.5, min_tracking_confidence=0.5)
```

```python
image = cv2.imread('face.jpg')
# Convert it to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# img = mpimg.imread('test.jpg')
imgplot = plt.imshow(image)
plt.show()
print(image.shape)



```

```python
# Pass image to holistic and get results
holistic_results = holistic.process(image)


```

```python
# Draw face, pose, and hand landmarks on original image
if holistic_results.face_landmarks:
  mp_drawing.draw_landmarks(
      image,
      holistic_results.face_landmarks,
      mp_holistic.FACE_CONNECTIONS,
      landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=2),
      connection_drawing_spec=mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2))

if holistic_results.pose_landmarks:
  mp_drawing.draw_landmarks(
      image,
      holistic_results.pose_landmarks,
      mp_holistic.POSE_CONNECTIONS,
      landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=2),
      connection_drawing_spec=mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2))

if holistic_results.left_hand_landmarks:
  mp_drawing.draw_landmarks(
      image,
      holistic_results.left_hand_landmarks,
      mp_holistic.HAND_CONNECTIONS,
      landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=2),
      connection_drawing_spec=mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2))

if holistic_results.right_hand_landmarks:
  mp_drawing.draw_landmarks(
      image,
      holistic_results.right_hand_landmarks,
      mp_holistic.HAND_CONNECTIONS,
      landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=2),
      connection_drawing_spec=mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2))


```
